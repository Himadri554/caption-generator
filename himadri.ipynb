```
import gradio as gr
from transformers import pipeline
import random

caption_generator=pipeline("text-generation",model="gpt2")
```

2.Challenge

Ethical concerns: Misleading audiences into thinking content is entirely human generated can erode trust.

2.Solution

Transparency: Be upfront about using AI-generated content to avoid misleding you audience.


```
emoji_dict={
    "positive": ["ðŸ˜Š","ðŸŒŸ","ðŸ”¥","ðŸ’ª","ðŸš€","âœ¨"],
    "negative":["ðŸ˜¢","ðŸ˜ž","ðŸ’”","ðŸ˜ ","ðŸ˜“"],
    "neutral":["ðŸ™‚","ðŸ˜","ðŸ§","ðŸ¤”","ðŸ˜¶"]
}


sentiment_pipeline=pipeline("sentiment-analysis")


def get_emojis(text):
  label=sentiment_pipeline(text)[0]['label'].lower()
  return''.join(random.sample(emoji_dict.get(label,["ðŸ˜"]),3))
```

3.Challenge

Over-Reliance on Emojis: Excessive use of emojis can make captions seem inauthentic.

3.Solution

Mindful Emoji Use:Use emojis judiciously and avoid overusing them , as it can make captions seem inauthentic.


```
def get_hashtags(prompt, platform):
  words=prompt.lower().split()
  tags=["#" + word.replace(" ","")for word in words if len(word)>3]


  platform_tags={
      "Instagram":["#instadaily","#igers","#pipcoftheday"],
      "Youtube":["#videos","#content creator","#videography"],
      "Snapchat":["#snaps","#story","#streak"]
  }
  return" ".join(tags[:5]+ random.sample(platform_tags[platform],2))
```

```
def generate_post(prompt, platform):
  #Generate caption
  caption = caption_generator(prompt, max_length=50, num_return_sequences=1)[0]['generated_text']


  #Generate emojis
  emojis=get_emojis(caption)


  #Generate hashtags
  hashtags = get_hashtags(prompt, platform)


  return caption.strip(), emojis, hashtags
```

4.Challenge

Platform specificity: Different social media platforms have varying formatting requirements and best practices for captions.


4.Solution

platform optimization : Tailor your captions to the specific platforms by considering length,formatting and tone.

```
interface = gr.Interface(
    fn=generate_post,
    inputs=[
         gr.Textbox(label="Enter keyword or theme"),
        gr.Radio(["Instagram", "Youtube", "Snapchat"],label="Choose Platform")
        ],
    outputs=[
        gr.Textbox(label="Generated Caption"),
        gr.Textbox(label="Emojis"),
        gr.Textbox(label="Hashtags")
    ],
    title="Social Media Post & Caption Generator",
    description="Generate catchy captions, relevent hashtags, and emojis based on your theme!"
    )



interface.launch()
```
